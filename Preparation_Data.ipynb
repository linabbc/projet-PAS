```python
import numpy as np
import pandas as pd

from Tools import preprocessing
from Tools import Vectoriser

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

On commence par importer le jeu de données.

```python
data = pd.read_csv("./Data/questions réponses PAS.txt", sep = "|")
```

Nous préparons les questions en leurs faisant passer les différentes étapes classique de pré-processing pour des données textuelles.( retirer les accents, les majuscules, stemmatisation ... )

```python
data["Question_Origine"] = data["Question"]
```

```python
data["Question"] = preprocessing(data["Question"])
```

```python
data["Questions_Lem"] = data["Question"].apply(lambda word : [" ".join(word)])
```

Nous définissons les phrases traitées

```python
sentences = data["Questions_Lem"].sum()
```

Nous les vectorisons de deux manières différentes. 

```python
data = Vectoriser(CountVectorizer(), data, sentences)
```

```python
data = Vectoriser(TfidfVectorizer(), data, sentences)
```

```python
data
```

Nous exportons le dataset.

```python
data.to_pickle("./Data/Dataset.pkl")
```

Et le corpus de mot

```python
with open("corpus.txt", "w") as writer:
    for sentence in sentences:
        writer.write(sentence+"\n")
```
